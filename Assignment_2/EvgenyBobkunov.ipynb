## Task 2. Self-supervised learning [12pts]

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model

# Load CIFAR-10 dataset
(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()

# Normalize pixel values between 0 and 1
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Define autoencoder architecture
input_img = Input(shape=(32, 32, 3))

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)

# Decoder
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

# Define autoencoder model
autoencoder = Model(input_img, decoded)

# Compile the model
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the autoencoder
autoencoder.fit(x_train, x_train, epochs=50, batch_size=128, shuffle=True, validation_data=(x_test, x_test))

# Save the model
autoencoder.save('autoencoder_model_bob.keras')

```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize pixel values between 0 and 1
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Split the dataset into training and validation sets
x_train, _, y_train, _ = train_test_split(x_train, y_train, train_size=0.1, stratify=y_train, random_state=42)

# Define autoencoder architecture
input_img = Input(shape=(32, 32, 3))

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
encoder_output = MaxPooling2D((2, 2), padding='same')(x)

# Define autoencoder model
autoencoder = Model(input_img, encoder_output)

# Freeze the weights of the encoder
for layer in autoencoder.layers:
    layer.trainable = False

# Classification head
x = Flatten()(encoder_output)
x = Dense(128, activation='relu')(x)
output = Dense(10, activation='softmax')(x)

# Define classification model
classification_model = Model(input_img, output)

# Compile the model
classification_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the classification model
history = classification_model.fit(x_train, y_train, epochs=20, batch_size=128, validation_data=(x_test, y_test))

# Evaluate the model on the entire test set
test_loss, self_supervised_test_acc = classification_model.evaluate(x_test, y_test)
print("Test Accuracy:", self_supervised_test_acc)

# Number of labeled data
self_supervised_num_labeled_data = len(x_train)
print("Used number of labeled data:", self_supervised_num_labeled_data)

# Number of unlabeled data
self_supervised_num_unlabeled_data = len(x_train) - self_supervised_num_labeled_data
print("Used number of unlabeled data:", self_supervised_num_unlabeled_data)

# Total size of the model
classification_model.summary()

# Inference time
start_time = time.time()
predictions = classification_model.predict(x_test[:1])  # Predict for a single image
end_time = time.time()
self_supervised_inference_time_per_image = end_time - start_time
print("Inference time per image:", self_supervised_inference_time_per_image, "seconds")

# Plot train and test losses
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot train and test accuracies
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

## Task 3. Auxiliary learning [12 pts]

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Dense
from tensorflow.keras.models import Model

# Define architecture
input_img = Input(shape=(32, 32, 3))

# Shared encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

# Classification branch
x = Flatten()(encoded)
x = Dense(128, activation='relu')(x)
classification_output = Dense(10, activation='softmax', name='classification')(x)

# Decoder for reconstruction
x = Conv2D(64, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

# Define model
model = Model(inputs=input_img, outputs=[classification_output, decoded])

# Assuming all data is labeled
num_labeled_data = len(x_train)

# Total size of the model
model.summary()

# Define loss functions
classification_loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
reconstruction_loss_fn = tf.keras.losses.MeanSquaredError()

# Define loss weights
classification_loss_weight = 0.7
reconstruction_loss_weight = 0.3

# Compile the model
model.compile(optimizer='adam',
              loss={'classification': classification_loss_fn},
              loss_weights={'classification': classification_loss_weight},
              metrics={'classification': 'accuracy'})

# Train the model
history = model.fit(x_train, {'classification': y_train},
                    epochs=20, batch_size=128, validation_data=(x_test, {'classification': y_test}))

# Evaluate the model on the entire test set
test_loss, test_classification_accuracy = model.evaluate(x_test, {'classification': y_test})
print("Test Classification Loss:", test_loss)
print("Test Classification Accuracy:", test_classification_accuracy)

# Inference time per image
start_time = time.time()
predictions = model.predict(x_test[:1])  # Predict for a single image
end_time = time.time()
aux_inference_time_per_image = end_time - start_time
print("Inference time per image:", aux_inference_time_per_image, "seconds")

# Plot train and test losses for classification task
plt.plot(history.history['loss'], label='Train Classification Loss')
plt.plot(history.history['val_loss'], label='Test Classification Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot train and test accuracies for classification task
plt.plot(history.history['classification_accuracy'], label='Train Classification Accuracy')
plt.plot(history.history['val_classification_accuracy'], label='Test Classification Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
